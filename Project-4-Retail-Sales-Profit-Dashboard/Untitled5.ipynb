```python
discount_amount = (df['List Price'] * df['Discount Percent'])/100
```

```python
discount_amount.head()
```

```python
df['Discount Amount'] = discount_amount
```

```python
df.head()
```

```python
df.columns
```

```python
df['Selling Price'] = df['List Price'] - df['Discount Amount']
```

```python
df['Profit'] = (df['Selling Price'] - df['cost price']) * df['Quantity']
```

```python
df.head()
```

```python
df['Profit'].sum()
```
